```python
# loading dataset using pandas
import pandas as pd

df = pd.read_csv('ecommerce_10000.csv')
```

```python
df.head(10)
```

```python
df.info()
```

```python
# Summary of Statistics using .describe()
df.describe(include='all')
```

```python
# Checking if missing data or null data present in the dataset
df.isnull().sum()
```

```python
df.columns
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns
```

```python
# Correcting Data columns from str (dtype) to datatime using pandas
df["orderdate"] = pd.to_datetime(df["orderdate"], format='%d-%m-%Y')
```

```python
df.info()
```

```python
# checking brand consistency whether having like 'Apple', 'apple'
df['brand'].unique()
```

```python
# Checking impossible values
df[df['rating'] > 5 ]
```

```python
# Creating Month and Year for insights
df["month"] = df["orderdate"].dt.month
df["year"] = df["orderdate"].dt.year
df["day_name"] = df["orderdate"].dt.day_name()
```

```python
df = df.rename(columns={'totalamount':'total_amount'})
df = df.rename(columns={'orderdate':'order_date'})
```

```python
df.head(3)
```

```python
df.to_csv("cleaned_ecommerce.csv", index=False)
```
